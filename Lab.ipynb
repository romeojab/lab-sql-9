### 1. How many distinct (different) actors' last names are there?

```python
# Literal understanding of the question
query = '''SELECT COUNT(DISTINCT(last_name)) 
        FROM actor;'''
data = pd.read_sql_query(query, engine)
data
```

```python
# I think the question is related to how many unqiue last name, so last names that are not shared with another actor
query1 = '''SELECT	last_name
        , COUNT(last_name) AS count_of_l_name
FROM actor
GROUP BY last_name
HAVING count_of_l_name = 1;'''
data = pd.read_sql_query(query1, engine)
len(data)
```

### 2. Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week

```python
query2 = '''SELECT	*
		, CASE
			WHEN DATE_FORMAT(rental_date, '%%w') IN (6,0) THEN 'Weekend'
            ELSE 'Workday'
		 END AS Day_type
            
FROM rental;'''
data = pd.read_sql_query(query2, engine)
data
```

### 3. Get all films with ARMAGEDDON in the title

```python
query3 = 'SELECT * FROM film WHERE title LIKE %s;'
data = pd.read_sql_query(query3, engine, params = ('%ARMAGEDDON%',))
data
```

### 4. Get 10 the longest films

```python
query4 = 'SELECT * FROM film ORDER BY length desc LIMIT 10;'
data = pd.read_sql_query(query4, engine)
data
```

### 5. How many films include Behind the Scenes content?

```python
query5 = "SELECT COUNT(*) AS count_of_films_with_special_features FROM film WHERE special_features LIKE '%%Behind the Scenes%%';"
data = pd.read_sql_query(query5, engine)
data
```

### 6. Which kind of movies (rating) have a mean duration of more than two hours?

```python
query6 = "SELECT rating, ROUND(AVG(length),2) AS 'Mean Duration' FROM film GROUP BY rating HAVING `Mean Duration` > 120;"
data = pd.read_sql_query(query6, engine)
data
```

### 7. Rank films by length (filter out the rows that have nulls or 0s in length column). In your output, only select the columns title, length, and the rank

```python
query7 = "SELECT title, length, RANK() OVER (ORDER BY length) AS 'Film Rank' FROM film WHERE length IS NOT NULL AND length > 0;"
data = pd.read_sql_query(query7, engine)
data
```
